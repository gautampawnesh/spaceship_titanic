```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
train_process_df = pd.read_csv("../intermediate_storage/titanic_features.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
cols= ['RoomService', 'FoodCourt', 'ShoppingMall', 'VRDeck', "Age"]
scaler = StandardScaler().fit(train_process_df[cols])
```

```python
train_process_df[cols] = scaler.transform(train_process_df[cols])
```

```python
y_encoded = train_process_df["Transported"]
```

```python
y_encoded
```

```python
X_encoded = train_process_df.copy()
X_encoded.drop(columns=["Transported", "Unnamed: 0"], inplace=True)
```

```python
X_encoded.columns
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X_encoded, y_encoded)
```
