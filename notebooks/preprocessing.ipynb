#### Name

```python
train_process_df.Name.isnull().sum()
```

```python
import random
import string
```

```python
letters = string.ascii_lowercase
```

```python
def df_random_name(s):
    if s is np.nan:
        return"".join(random.choices(letters, k=10))
    else:
        return s
```

```python
train_process_df["Name"] = train_process_df.Name.apply(lambda x: df_random_name(x))
```
